2. Checking for Sample Representativity

As we mentioned earlier, most of the courses we offer are on web and mobile development, but we also cover many other domains, like data science, game development, etc. For the purpose of our analysis, we want to answer questions about a population of new coders that are interested in the subjects we teach. We'd like to know:

1. Where are these new coders located.
2. What are the locations with the greatest number of new coders.
3. How much money new coders are willing to spend on learning

Before starting to analyze the sample data we have, we need to clarify whether it's representative for our population of interest and it has the right categories of people for our purpose

```python
#we are going to generate the frequency distribution table for the JobRoleInterest column
survey_data['JobRoleInterest'].value_counts(normalize=True)*100
```

```python
#Eliminating nulls from the series and splitting the dataframe according to interests
interests_no_nulls = survey_data["JobRoleInterest"].dropna()
subject_of_interest = interests_no_nulls.str.split(",")
subject_of_interest
```

```python
#Accounting for the amount of interests for each person surveyed and putting it into a frequency table
options = subject_of_interest.apply(lambda x: len(x))
options_table = options.value_counts(normalize = True).sort_index()*100
print(options_table)
```

Observations:
1. There are some people who are interested in only 1 subject and others are interested in more than 1 subject
2. 31% of the population is interested in only 1 subject. The rest are interested in 2+ subjects. I still consider this sample to be representative. We will look how many are interested in web and mobile development

```python
#We will look into how many people are interested in web and movule development
web_or_mobile = interests_no_nulls.str.contains("Web Developer|Mobile Developer")
web_or_mobile_table = web_or_mobile.value_counts(normalize=True)*100
print(web_or_mobile_table)
#We will graph the results to get a better understanding
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")
web_or_mobile_table.plot.bar()
plt.title("86% of the participants are interested in Web or Mobile Development", y= 1.08)
plt.xticks([0,1],["Web/Mobile development", "Other subjects"], rotation = 0)
plt.ylabel("Percentage (%)")
plt.show()
```

86% of the participants have interests in either web or mobile development

3. New Coders - Locations and Densities

Now that we found out that the sample has the right categories of people for our purpose, we can begin analyzing it. We can start with finding out where these new coders are located, and what are the densities (how many coders there are) for each location.

The data set provides information about the location of each participant at a country level. The CountryCitizen variable describes the country of origin for each participant, and the CountryLive variable describes what country each participants lives in (which may be different than the origin country).

For our analysis, we'll work with the CountryLive variable because we're interested where people actually live at the moment when we run the ads. In other words, we're interested where people are located, not where they were born.

Because the data set provides information at a country level, we can think of each country as an individual market. This means we can frame our goal as finding the two best countries to advertise in.

One indicator of a good market is the number of potential customers — the more potential customers in a market, the better. If our ads manage to convince 10% of the 5000 potential customers in market A to buy our product, then this is better than convincing 100% of the 30 potential customers in market B

```python
#Cleaning the data by leaving out null records in the Job Role Interest field
survey_data_clean = survey_data[survey_data["JobRoleInterest"].notnull()].copy()
freq_table_country_abs = survey_data_clean["CountryLive"].value_counts()
freq_table_country_rel = survey_data_clean["CountryLive"].value_counts(normalize = True)*100
#print (freq_table_country_abs)
#print (freq_table_country_rel)
pd.DataFrame(data = {"absolute": freq_table_country_abs, "relative": freq_table_country_rel})
```

Based on the results it seems that the most interesting markets are USA (45%) and India (7.7%)

4. Spending Money for Learning

Previously, we found useful information about the location of new coders, and what are the countries where new coders live, but we need to go more in depth with our analysis before taking a decision. We can continue by figuring out how much money new coders are actually willing to spend on learning. Advertising within markets where most people are only willing to learn for free is extremely unlikely to be profitable for us.

The MoneyForLearning column describes in American dollars the amount of money spent by participants from the moment they started coding until the moment they completed the survey. Our company sells subscriptions at a price of $59 per month, and for this reason we're interested in finding out how much money each student spends per month.

It also seems like a good idea to narrow down our analysis to only four countries: the US, India, the United Kingdom, and Canada. Two reasons for this decision are:

These are the countries having the highest absolute frequencies in our sample, which means we have a decent amount of data for each.
Our courses are written in English, and English is an official language in all these four countries. The more people that know English, the better our chances to target the right people with our ads.

```python
#Replacing students that recorded 0 months of programming with 1
survey_data_clean["MonthsProgramming"].replace(0,1,inplace = True)
#Creating column that calculates the amount of money each student spends per month
survey_data_clean["SpentPerMonth"] = survey_data_clean["MoneyForLearning"]/survey_data_clean["MonthsProgramming"]
survey_data_clean["SpentPerMonth"].isnull().sum()
```

```python
#Removing missing values from DF according to money spent and the country they live in
survey_data_clean = survey_data_clean[survey_data_clean["SpentPerMonth"].notnull()]
survey_data_clean = survey_data_clean[survey_data_clean["CountryLive"].notnull()]
```

```python
#Grouping by country to find out how much is spent monthly per student per country
countries_mean = survey_data_clean.groupby("CountryLive")["SpentPerMonth"].mean().sort_index()
countries_mean[["United States of America","India","United Kingdom","Canada"]]
```

The average monthly spenditure for the UK seem way off. It seems we need to understand better the data to account for this behavior

6· Dealing with Extreme Outliers

Out of the 11 outliers from USA we can observe that 6 attended a bootcamp. For the others it is hard to infer why did they spent that amount of money. Consequently only 8 of the outliers have only been programming less than 3 months, so the amount of monthly spenditure is probably for a long term coding program or bootcamp. For these reasons we will eliminate these records 8 records from the analysis

```python


#Generating box plots for the mean in average spenditure to identify outliers
only_4 = survey_data_clean[survey_data_clean["CountryLive"].str.contains("United States of America|India|United Kingdom|Canada")]
import seaborn as sns
sns.boxplot (x="CountryLive",y="SpentPerMonth", data=only_4)
```

```python
#Removing outliers above $20Kb
only_4 = survey_data_clean[survey_data_clean["CountryLive"].str.contains("United States of America|India|United Kingdom|Canada")]
import seaborn as sns
sns.boxplot (x="CountryLive",y="SpentPerMonth", data=only_4)
```

The results for the United Kingdom and Canada are surprisingly low relative to the values we see for India. If we considered a few socio-economical metrics (like GDP per capita), we'd intuitively expect people in the UK and Canada to spend more on learning than people in India.

It might be that we don't have have enough representative data for the United Kingdom, Canada, and India, or we have some outliers (maybe coming from wrong survey answers) making the mean too big for India, or too low for the UK and Canada. Or it might be that the results are correct.

```python
#Analyzing India's Outliers
india_outliers = only_4[(only_4["CountryLive"] == "India") & (only_4["SpentPerMonth"]>2500)]
india_outliers
```

It appears that there is no special reason to include Indias outliers in th analysis, since none of them attended an extended bootcamp to justify the cost.


```python
#Dropping the India outliers by using the row index
only_4 = only_4.drop(india_outliers.index)
```

```python

#Analyzing USA's outliers
USA_outliers = only_4[(only_4["CountryLive"] == "United States of America")&(only_4["SpentPerMonth"] >=6000)]
USA_outliers
```

```python
#Dropping USA outliers that did not attend a bootcamp
no_bootcamp = only_4[(only_4["AttendedBootcamp"] == 0) & (only_4["CountryLive"]=="United States of America") & (only_4["SpentPerMonth"]>=6000)]
only_4 = only_4.drop(no_bootcamp.index)
```

```python
less_3_months = only_4[(only_4["MonthsProgramming"] <= 3) & (only_4["CountryLive"]=="United States of America") & (only_4["SpentPerMonth"]>=6000)]
only_4 = only_4.drop(less_3_months.index)
```

```python

#Analyzing Canada outliers
Canada_outliers = only_4[(only_4["CountryLive"] == "Canada")&(only_4["SpentPerMonth"]>4000)]
Canada_outliers
```

The only Canada outlier also attended a bootcamp, so his monthly spenditure is unrealistic. For this reason we will eliminate this record

```python
#Dropping Canada outlier
only_4 = only_4.drop(Canada_outliers.index)
only_4.groupby("CountryLive").mean()["SpentPerMonth"]
```

```python
sns.boxplot (x="CountryLive",y="SpentPerMonth", data=only_4)
plt.title ("Money Spent per Month per Country")
plt.ylabel("Monthly Spenditure")
plt.xlabel("Country")
```

The best markets to focus our advertising efforts, according to average monthly spenditure, are USA and Canada. But lets look a little bit deeper into the data and see if India is still worthwhile
